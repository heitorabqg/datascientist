```python
# gerar funo de wiener

def Wiener_HA(x: np.ndarray, d: np.ndarray, M: int):
  R = np.zeros((M,M))
  p = np.zeros(M)
  p = p.reshape(-1,1)

  x_M = np.zeros(M)
  x_M = x_M.reshape(-1,1)

  # numero de amostras
  N = x.shape[0]

  for index_W in range (N):
      # vetor x dentro do loop

      x_M[1:M] = x_M[0:M-1]
      x_M[0] = x[index_W]

#      print("Valores de x_M \n", x_M)

      # Tamanho de R (Nx1) (1xN) = NxN
      R = R + x_M@x_M.T

#      print("Valores de R \n", R)
      p = p + x_M * d[index_W]

#      print("Valores de p \n", p)

  R = R/N
  p = p/N

  # o=1

  w_o = np.linalg.solve(R, p)

  return w_o, R, p

```
