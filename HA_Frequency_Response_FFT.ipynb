```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes do filtro FIR
h = np.array([0.3089, -0.1685])

# Frequncias normalizadas de 0 a pi (em rad/sample)
w = np.linspace(0, np.pi, 512)

# Resposta em frequncia
H = np.fft.fft(h, 512)
H = H[:4096]  # Considera apenas a metade positiva (de 0 a pi)
w = w[:4096]

# Magnitude e fase
magnitude = np.abs(H)
phase = np.angle(H)

# Plotando
plt.figure(figsize=(12, 5))

# Magnitude
plt.subplot(1, 2, 1)
plt.plot(w, magnitude)
plt.title('Resposta em Magnitude')
plt.xlabel('Frequncia (rad/amostra)')
plt.ylabel('|H(e^{j})|')
plt.grid(True)

# Fase
plt.subplot(1, 2, 2)
plt.plot(w, phase)
plt.title('Resposta em Fase')
plt.xlabel('Frequncia (rad/amostra)')
plt.ylabel('H(e^{j}) [rad]')
plt.grid(True)

plt.tight_layout()
plt.show()

kindex = int(512/5)
print("Indice k em 2pi/10 ", kindex)
print("Phase em k = 10 em rad", phase[kindex])
print("Phase em k = 10 em graus", phase[kindex]*180/np.pi)
print("Magnitude em k = 10", magnitude[kindex])

print("Phase do sinail interferente em rad", np.pi/6)
```
