# MLP genérica para solução do problemas das meias luas

Neste exercício vamos treinar uma rede MLP com o algorotimo backpropagation para a solução do problema das meias luas.

````python
import numpy as np
import matplotlib.pyplot as plt
````

Vamos gerar os dados de treinamento do problema das meias-luas.

````python
def meias_luas(NA, NB, r1, r2, r3):
    """
    dados = meias_luas(NA,NB,r1,r2,r3)
    NA: número de pontos da região A
    NB: número de pontos da região B
    r1, r2 e r3: dados das meias-luas
    """

    # total de dados de treinamento
    Nt = NA + NB

    # dados das meia luas
    rmin = r1 - r3 / 2
    rmax = r1 + r3 / 2

    # Pontos da Região A
    a = np.pi * np.random.rand(NA, 1)
    rxy = np.random.uniform(rmin, rmax, (NA, 1))
    x1A = rxy * np.cos(a)
    x2A = rxy * np.sin(a)
    dA = np.ones((NA, 1))
    pontosA = np.hstack((x1A, x2A, dA))

    # Pontos da Região B
    a = np.pi * np.random.rand(NB, 1)
    rxy = np.random.uniform(rmin, rmax, (NB, 1))
    x1B = rxy * np.cos(a) + r1
    x2B = -rxy * np.sin(a) - r2
    dB = -np.ones((NB, 1))
    pontosB = np.hstack((x1B, x2B, dB))

    # Concatenando e embaralhando os dados
    dados = np.vstack((pontosA, pontosB))
    np.random.shuffle(dados)

    # Figura para mostrar os dados de treino
    fig, ax1 = plt.subplots()
    ax1.plot(x1A, x2A, ".b")
    ax1.plot(x1B, x2B, ".r")
    plt.xlabel("x_1")
    plt.ylabel("x_2")
    plt.grid(axis="x", color="0.5")
    plt.grid(axis="y", color="0.5")

    return dados
````

````python
# número de pontos de treinamento da Região A
NA = 500

# número de pontos de treinamento da Região B
NB = 500

# número total de dados de treinamento
Nt = NA + NB

r1 = 10
r3 = 6
r2 = -4

dados_treino = meias_luas(NA, NB, r1, r2, r3)
````

Após implementação da função para treinamento, treine uma rede para solucionar o problemas das meias luas considerando os dados `x`e `d` criados anteriormente. Faça um gráfico da função custo ao longo das épocas de treinamento.

## Resolução

````python
import numpy as np

def redeMLP_SGD(x, d, Nn, eta, Nt, Nb, Ne, W0=None):
    """
    Treinamento de uma rede MLP usando o algoritmo Backpropagation (modo Mini-Batch)

    Entradas:
    ----------
    x : np.ndarray
        Sinal de entrada (Nt x Nn[0])
    d : np.ndarray
        Sinal desejado (Nt x Nn[-1])
    Nn : list
        Estrutura da rede. Ex: [2, 3, 2, 1]
    eta : float
        Passo de adaptação (taxa de aprendizado)
    Nt : int
        Número de dados de treinamento
    Nb : int
        Tamanho do mini-batch (número de amostras por iteração)
    Ne : int
        Número de épocas de treinamento
    W0 : list[np.ndarray] ou None
        Pesos iniciais. Se None, os pesos são inicializados aleatoriamente.
        Cada W[j] tem dimensão (Nn[j], Nn[j-1] + 1)

    Saídas:
    -------
    J_MSE : np.ndarray
        Valor da função custo (MSE) ao longo das épocas
    W : list[np.ndarray]
        Pesos (incluindo bias) da última época
    """

    # Número de camadas (sem contar a de entrada)
    L = len(Nn) - 1

    # Inicialização dos pesos
    if W0 is None:
        W = [np.random.uniform(-0.5, 0.5, (Nn[j], Nn[j-1] + 1)) for j in range(1, len(Nn))]
    else:
        W = [np.copy(W0[j]) for j in range(len(W0))]

    # Função de ativação e derivada
    f = lambda x: np.tanh(x)
    f_deriv = lambda x: 1 - np.tanh(x)**2

    # Vetor para armazenar o erro quadrático médio por época
    J_MSE = np.zeros(Ne)

    # Loop de treinamento
    for epoca in range(Ne):
        soma_erro = 0

        # Embaralha os índices dos dados
        idx = np.random.permutation(Nt)
        x = x[idx, :]
        d = d[idx, :]

        # Loop de Mini-Batch
        for i in range(0, Nt, Nb):
            # Seleciona o mini-batch
            x_batch = x[i:i+Nb, :]
            d_batch = d[i:i+Nb, :]

            # Inicializa os gradientes para os mini-batches
            gradientes = [np.zeros_like(Wj) for Wj in W]

            # Processa o mini-batch (Forward + Backward pass)
            for n in range(len(x_batch)):
                # ----- Forward pass -----
                y = [None] * (L + 1)
                v = [None] * (L + 1)

                y[0] = x_batch[n, :].reshape(-1, 1)  # entrada coluna

                for j in range(1, L + 1):
                    x_j = np.vstack(([1], y[j-1]))  # adiciona bias
                    v[j] = np.dot(W[j-1], x_j)
                    y[j] = f(v[j])

                # Erro na saída
                e = d_batch[n, :].reshape(-1, 1) - y[L]
                soma_erro += np.mean(e**2)

                # ----- Backward pass -----
                delta = [None] * (L + 1)

                # Camada de saída
                delta[L] = f_deriv(v[L]) * e

                # Demais camadas
                for j in range(L-1, 0, -1):
                    W_no_bias = W[j][:, 1:]  # remove pesos do bias
                    delta[j] = f_deriv(v[j]) * np.dot(W_no_bias.T, delta[j+1])

                # Acumula gradientes para o mini-batch
                for j in range(1, L + 1):
                    x_j = np.vstack(([1], y[j-1]))  # inclui bias
                    gradientes[j-1] += eta * np.dot(delta[j], x_j.T)

            # Atualiza os pesos após o mini-batch
            for j in range(1, L + 1):
                W[j-1] += gradientes[j-1] / Nb

        # Calcula o MSE da época
        J_MSE[epoca] = soma_erro / Nt

        #print(f"Epoca {epoca+1}/{Ne} - MSE = {J_MSE[epoca]:.6f}")

    plt.figure(figsize=(7,5))
    plt.plot(J_MSE, 'b-', linewidth=2)
    plt.title('Evolução da Função Custo (MSE) durante o Treinamento', fontsize=12)
    plt.xlabel('Epoca', fontsize=11)
    plt.ylabel('J_MSE', fontsize=11)
    plt.grid(True)
    plt.show()

    return J_MSE, W
````

## Resolução

````python
import numpy as np

def redeMLP_teste(x, d, W, Nn, Nteste):
    """
    Teste da rede MLP treinada (modo feedforward)

    Entradas:
    ----------
    x : np.ndarray
        Sinal de entrada (Nteste x Nn[0])
    d : np.ndarray
        Sinal desejado (Nteste x Nn[-1])
    W : list[np.ndarray]
        Lista de matrizes de pesos (incluindo bias) da última época
    Nn : list[int]
        Estrutura da rede, ex: [2, 5, 1]
    Nteste : int
        Número de amostras de teste

    Saídas:
    -------
    J_MSE : float
        Erro quadrático médio no teste
    y : np.ndarray
        Saída da rede para cada amostra (Nteste x Nn[-1])
    """

    L = len(Nn) - 1  # número de camadas ocultas + saída
    f = lambda x: np.tanh(x)  # função de ativação

    y = np.zeros((Nteste, Nn[-1]))

    soma_erro = 0

    for n in range(Nteste):
        # Forward pass
        yj = x[n, :].reshape(-1, 1)
        for j in range(1, L + 1):
            xj = np.vstack(([1], yj))  # inclui o bias
            vj = np.dot(W[j-1], xj)
            yj = f(vj)
        y[n, :] = yj.T

        # Cálculo do erro quadrático
        e = d[n, :].reshape(-1, 1) - yj
        soma_erro += np.mean(e**2)

    J_MSE = soma_erro / Nteste
    return J_MSE, y
````
