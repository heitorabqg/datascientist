```python
import numpy as np
from scipy.special import erfinv

# Example: Calculate the inverse Q-function for a probability x
x = 0.0005  # A probability value between 0 and 1
inverse_q_value = np.sqrt(2) * erfinv(1 - 2 * x)
print(f"The inverse Q-function of {x} is: {inverse_q_value}")
```
